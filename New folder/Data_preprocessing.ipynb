```python
import pandas as pd

df = pd.read_csv('1) iris.csv')

print("Dataset ki shuruati rows:")
print(df.head())

print("\nData Information:")
print(df.info())
```

```python
print("Missing values ki sankhya:")
print(df.isnull().sum())

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['species'] = le.fit_transform(df['species'])

print("Encoding ke baad data (0, 1, 2 represent species):")
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('species', axis=1)
y = df['species']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled Features (Pehli 5 rows):")
print(X_scaled[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training set ka size: {X_train.shape[0]} rows")
print(f"Testing set ka size: {X_test.shape[0]} rows")
print("\nTask 1: Preprocessing complete ho gaya hai!")
```
